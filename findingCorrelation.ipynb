```python
#FLOOD DATA (People Affected by Flooding)
col_names = ["country","1970","1971","1972","1973","1974","1975","1976","1977","1978","1979","1980","1981","1982","1983","1984","1985","1986","1987","1988","1989","1990","1991","1992","1993","1994","1995","1996","1997","1998","1999","2000","2001","2002","2003","2004","2005","2006","2007","2008",""]
url = 'https://raw.githubusercontent.com/mgee18/datascience/master/csvFiles/flood_affected.csv'
floodData = pd.read_csv(url, header=0, names=col_names, sep=',')
floodData.drop(floodData.columns[[x for x in range(36,41)]], axis=1, inplace=True)
floodData.head()
```

```python
floodDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for flood_country_index in range(0, len(floodData.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == floodData.ix[flood_country_index,0]:
            countryFound = True
            for x in range(0, 10):
                floodDataList.append(np.nan)
            for year in [str(x) for x in range(1970, 2000)]:
                floodDataList.append(floodData[year][flood_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            floodDataList.append(np.nan)
    countryFound = False
```

```python
GDPdataList = []
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for year in [str(x) for x in range(1960, 2000)]:
        GDPdataList.append(GDPdata[year][gdp_country_index])
```

```python
agDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for ag_country_index in range(0, len(agData.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == agData.ix[ag_country_index,0]:
            countryFound = True
            for year in [str(x) for x in range(1960, 2000)]:
                agDataList.append(agData[year][ag_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            agDataList.append('')
    countryFound = False
print(len(agDataList))
```

```python
SAdataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for sa_country_index in range(0, len(SAdata.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == SAdata.ix[sa_country_index,0]:
            countryFound = True
            for year in [str(x) for x in range(1960, 2000)]:
                SAdataList.append(SAdata[year][sa_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            SAdataList.append('')
    countryFound = False
print(len(SAdataList))
```

```python
popDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for pop_country_index in range(0, len(agData.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == popData.ix[pop_country_index,0]:
            countryFound = True
            for year in [str(x) for x in range(1960, 2000)]:
                popDataList.append(popData[year][ag_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            popDataList.append('')
    countryFound = False
print(len(popDataList))
```

```python
upopDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for upop_country_index in range(0, len(upopData.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == upopData.ix[upop_country_index,0]:
            countryFound = True
            for year in [str(x) for x in range(1960, 2000)]:
                upopDataList.append(upopData[year][upop_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            upopDataList.append('')
    countryFound = False
print(len(upopDataList))
```

```python
oilDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for oil_country_index in range(1, len(oilData.ix[:,0]) + 1):
        if GDPdata.ix[gdp_country_index,0] == oilData.ix[oil_country_index,0]:
            countryFound = True
            oilDataList.append(0.0)
            for year in [str(x) for x in range(1961, 2000)]:
                oilDataList.append(oilData[year][oil_country_index])
    if countryFound != True:
        for x in range(1960, 2000):
            oilDataList.append(0.0)
    countryFound = False

print(len(oilDataList))
```

```python
epidemicDataList = []
countryFound = False
for gdp_country_index in range(0, len(GDPdata.ix[:,0])):
    for epidemic_country_index in range(0, len(epidemicData.ix[:,0])):
        if GDPdata.ix[gdp_country_index,0] == epidemicData.ix[epidemic_country_index,0]:
            for x in range(0, 10):
                epidemicDataList.append('')
            countryFound = True
            for year in [str(x) for x in range(1970, 2000)]:
                if year != '1973':
                    epidemicDataList.append(epidemicData[year][epidemic_country_index])
                else:
                    epidemicDataList.append('')
    if countryFound != True:
        for x in range(1960, 2000):
            epidemicDataList.append('')
    countryFound = False
print(len(epidemicDataList))
```

```python
columns = {'GDP Data': GDPdataList, 'Flood Data': floodDataList, 'Agricultural Data': agDataList, 'Surface Area Data': SAdataList, 'Population Data': popDataList, 'Urban Population Data': upopDataList, 'Oil Data': oilDataList, 'Epidemic Data': epidemicDataList}
```

```python
data = pd.DataFrame.from_dict(columns)
```
